```python
def solution(numer1, denom1, numer2, denom2):
    answer = []
    
    numer = numer1*denom2 + numer2 * denom1
    denom = denom1 * denom2
    
    a, b = numer, denom

    # 약분
    while(b):
        a, b = b, a % b
        
    numer //= a
    denom //= a
    
    answer = [numer, denom]
    return answer
```

```python
# 올바른 괄호
def solution(s):
    answer = True
    stack = []
    for char in s:
        if char == '(':
            stack.append(char)
        elif char == ')':
            if not stack:
                answer = False
                break
            stack.pop()

    if stack:
        answer = False

    return answer
```

```python
def solution(n, m):
    answer = []
    # 최대공약수
    for i in range(min(n, m), 0, -1):
        if n % i == 0 and m % i == 0:
            answer.append(i)
            break
    # 최소공배수
    for i in range(max(n, m), (n * m) + 1):
        if i % n == 0 and i % m == 0:
            answer.append(i)
            break
    return answer

solution(3, 12)
solution(2, 5)
```

```python
def solution(n):
    answer = 0
    i = 1
    while i != n:
        sum = 0
        for j in range(i, n):
            sum += j
            if sum == n:
                answer += 1
            elif sum > n:
                break
        i += 1
    return answer + 1
```

```python
# 위 문제 투 포인터 알고리즘 사용
def solution(n):
    answer = 0
    start = 1
    current_sum = 0

    # end를 1부터 n까지 증가시키며 윈도우 확장
    for end in range(1, n + 1):
        current_sum += end

        # 합이 n보다 크면, 앞에서부터(start)
        while current_sum > n:
            current_sum -= start
            start += 1

        # 합이 정확히 n이 되면 카운트 증가
        if current_sum == n:
            answer += 1

    return answer
```

```python
def solution(s):
    answer = ''
    words = s.split(' ')
    for word in words:
        for i in range(len(word)):
            if i % 2 == 0:
                answer += word[i].upper()
            else:
                answer += word[i].lower()
        answer += ' '
    return answer[:-1]
    
```

```python
def solution(n):
    answer = 0
    binary_n  = bin(n)[2:]
    count = binary_n.count('1')
    while True:
        n += 1
        binary_n = bin(n)[2:]
        if binary_n.count('1') == count:
            answer = n
            break
    return answer

solution(78)
```
