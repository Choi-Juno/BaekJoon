```python
def solution(numer1, denom1, numer2, denom2):
    answer = []
    
    numer = numer1*denom2 + numer2 * denom1
    denom = denom1 * denom2
    
    a, b = numer, denom

    # 약분
    while(b):
        a, b = b, a % b
        
    numer //= a
    denom //= a
    
    answer = [numer, denom]
    return answer
```

```python
# 올바른 괄호
def solution(s):
    answer = True
    stack = []
    for char in s:
        if char == '(':
            stack.append(char)
        elif char == ')':
            if not stack:
                answer = False
                break
            stack.pop()

    if stack:
        answer = False

    return answer
```

```python
import math
# 정수 제곱근 판별
def solution(n):
    answer = 0
    if n % math.sqrt(n) == 0:
        answer = (math.sqrt(n) + 1) ** 2
    else:
        answer = -1

    return answer
```

```python
def solution(absolutes, signs):
    answer = 0
    for absolute, sign in zip(absolutes, signs):
        if sign:
            answer += absolute
        else:
            answer -= absolute

    return answer
```

```python
def solution(num):
    count = 0
    while num != 1:
        if num % 2 == 0:
            num /= 2
        else:
            num = num * 3 + 1
        count += 1
    if count >= 500:
        return -1
    if num == 1:
        return count

solution(6)
```

```python
def solution(arr):
    answer = []
    for i in range(len(arr)):
        if i == 0:
            answer.append(arr[i])
        elif arr[i] != arr[i-1]:
            answer.append(arr[i])
        else:
            continue
    return answer

solution([1,1,3,3,0,1,1])
```

```python
def solution(n, m):
    answer = []
    # 최대공약수
    for i in range(min(n, m), 0, -1):
        if n % i == 0 and m % i == 0:
            answer.append(i)
            break
    # 최소공배수
    for i in range(max(n, m), (n * m) + 1):
        if i % n == 0 and i % m == 0:
            answer.append(i)
            break
    return answer

solution(3, 12)
solution(2, 5)
```

```python
def solution(t, p):
    answer = 0
    for i in range(len(t) - len(p) + 1):
        if int(t[i:i+len(p)]) <= int(p):
            answer += 1
    return answer

solution("3141592", "271")
```
