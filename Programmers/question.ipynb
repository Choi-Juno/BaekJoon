```python
def solution(numer1, denom1, numer2, denom2):
    answer = []
    
    numer = numer1*denom2 + numer2 * denom1
    denom = denom1 * denom2
    
    a, b = numer, denom

    # 약분
    while(b):
        a, b = b, a % b
        
    numer //= a
    denom //= a
    
    answer = [numer, denom]
    return answer
```

```python
def solution(num):
    answer = 0
    count = [0] * 10001

    for i in num:
        count[i] += 1
    
    max_count = max(count)
    max_index = count.index(max_count)

    if count.count(max_count) > 1:
        return -1
    else:
        return max_index

solution(291183)
```

```python
# 올바른 괄호
def solution(s):
    answer = True
    stack = []
    for char in s:
        if char == '(':
            stack.append(char)
        elif char == ')':
            if not stack:
                answer = False
                break
            stack.pop()

    if stack:
        answer = False

    return answer
```
